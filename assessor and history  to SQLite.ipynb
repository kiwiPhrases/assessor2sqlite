```python
## load column headers and delineators
#assign file names
headerPaths = {}
headerPaths['history'] = "20130123_USC_HISTORY_LAYOUT RB_3 _032618.xls"
headerPaths['assessor'] = "DFS 3_0 Assessor Layout With Geocodes_032618.xls"

colDict = {}
folders = {}
#here, use the rbnames in the history file AND change to propid in the assossor file
#maybe add a loop with try upon faulty queries and just a loop for more than one query
#propid is unique for assessor
#transid is unique for history

#fetch the folders that contain the data files:
for key in ['history', 'assessor']:
    folders[key] = [s for s in os.listdir("/".join([data_path,'coreLogic'])) if ("." not in s) & (key in s.lower())]
    
#load and format header data    
colDict['history'] = pd.read_excel("/".join([data_path,'coreLogic',headerPaths['history']]), skiprows=3, header=0).iloc[:88]
#colDict['history'].loc[:,'fxdCols'] = colDict['history'].loc[:,'Field Name'].str.strip().str.replace("\s", "_").str.lower()
colDict['history'].loc[:,'fxdCols'] = colDict['history']['RB NAME'].str.strip()
colDict['history'].loc[87,'fxdCols'] = 'filler'
colDict['history'].fxdCols.replace("process_id", 'process_id_hist', inplace=True)

colDict['assessor'] = pd.read_excel("/".join([data_path,'coreLogic', headerPaths['assessor']]), skiprows = 1,header=0).iloc[:184]
colDict['assessor'].loc[:,'fxdCols'] = colDict['assessor'].loc[:,'Field Name'].str.replace("\s", "_").str.lower().str.replace("^mm_|^sa_",'')
colDict['assessor'].fxdCols.replace("property_id", 'propid', inplace=True)
colDict['assessor'].fxdCols.replace("process_id", 'process_id_ass', inplace=True)
```

## I. Read in the files and dump to sqlite database

*dates*: years are currently in floats which is not a bad thing. 
Since some are corrupt (ie 20010000), I will leave it in float so we just query year>=20010000

*process*: I read the files in chunks and dump each chunk into a database

### A. functions

```python
#single file
key='assessor'
def createInsertClause(key,chunk):
    insertClauses = chunk.apply(lambda row: " INSERT INTO %s VALUES "%key + str(tuple(row)) + ";",axis=1)
    return("BEGIN TRANSACTION; " + " ;".join(insertClauses) + " COMMIT;")

def processFile(fpath, key, chunkSize = 500000, dbname = dbname, histCols=None):
    hd = colDict[key]
    print("Working on: %s" %fpath)
    print("\tcreating file reader")
    file_start = time.time()
    reader = pd.read_fwf(fpath, widths = hd.Length.astype('int'), names = hd.fxdCols,header=None, chunksize = chunkSize)
    print("\tfile reader created")

    #dump file to sql in chunks
    for i,chunk in enumerate(reader):
        print("\tchunk num: %d" %i)
        start_time = time.time()
        chunk.drop('filler', axis=1, inplace=True)
        
        #set aside rows with propid==NULL into a separate table
        #then drop them from the main table
        nullRows = chunk.propid.isnull()
        if nullRows.sum()> 0:
            chunkNA  = chunk[chunk.propid.isnull()]
            print("\t\t\t[%d] rows with NA propid" %(nullRows.sum()))
        chunk.dropna(subset = ['propid'], inplace=True)

        chunk['propid'] = chunk.propid.astype(np.int64)

        if histCols is not None:
            chunk = chunk.loc[:,histCols]
        print("\tprocessing done")
        #print("\tcreating insert clause")
        #insertQuery = createInsertClause(key,chunk)
        print("\tinserting into table")
        #open database connection
        conn =  sqlite3.connect("/".join([data_path, dbname]))
        try:
            chunk.to_sql(key,conn, if_exists='append', index=False)
        except sqlite3.Error as er:
            print('\t\t',er)
            print('\t\tskipping chunk')
        conn.close()
        #c = conn.cursor()
        #c.executescript(insertQuery)
        #conn.commit()
        print("\tchunk took: %.2fsec" %(time.time()-start_time))
        print("\tfinal chunk size: %d" %chunk.shape[0])
        print("\t"+"--"*20)
    print("\tthis file took %.2fseconds" %(time.time() - file_start)) 
    #this seems necessary for some reason as data aren't dumped in the end
    #del chunk
    #del reader
    #close database connection
    
    print("done")

#convenience function to fetch the file name inside the folder
def fetchfile(subdirectory):
    parentPath = "/".join([data_path,'coreLogic' ,subdirectory])
    return("/".join([parentPath, os.listdir(parentPath)[0]]))

#iterate over folders:
def processFolders(key,folders,histCols=None):
    for folder in folders:
        processFile(fetchfile(folder),key, histCols=histCols)

#get table names    
def getTableNames(dbfile, lookfor="'table'"):
    #open connection
    conn =  sqlite3.connect("/".join([data_path, dbfile]))
    c = conn.cursor()

    #fetch names
    res = c.execute("SELECT name FROM sqlite_master WHERE type=%s;" %lookfor)
    print(res.fetchall())
    
    #close connection
    conn.close()    
    
def queryDB(query, dbfile):
    conn =  sqlite3.connect("/".join([data_path, dbfile]))
    print("Connection opened to [%s]" %dbfile)
    start_time = time.time()
    
    df = pd.read_sql_query(query, conn)
    
    print("\tDone. Query took: %.2fseconds" %(time.time()-start_time))
    print("\tsize of file read in:", df.shape)
    
    conn.close()
    print("\tConnection closed")
    return(df)
```

## DataQuick

There is some duplication in the property history records so in order to maintain a unique index `transid` in the history file, before dumping DataQuick history data to the database, I must first check which `transid` already exist, load them, drop them from the DataQuick file, then load into db. 

The other difference between CoreLogic's `processFile` and that of DataQuick is that the latter reads-in SAS files directly while the `processFile` for CoreLogic reads in tabulated files using two other files from which header and spacing information are read. 

```python
#modified file-processing function for dataquick
def processFile(fpath, key, uniqueID = 'propid',dropCols=None ,chunkSize = 1e5, dbname = dbname, histCols=None):
    print("Working on: %s" %fpath)
    #print("\tcreating file reader")
    reader = pd.read_sas(fpath,format = 'sas7bdat', chunksize=int(chunkSize))
    #print("\tfile reader created")
    
    
    start_time = time.time()
    
    #dump file to sql in chunks
    print("Commencing iterations on chunks")
    for i,chunk in enumerate(reader):
        print("chunk num: %d" %i)
        chunk_start = time.time()
        if dropCols is not None:
            chunk.drop(dropCols, axis=1, inplace=True)
        
        #set aside rows with propid==NULL into a separate table
        #then drop them from the main table
        print("\tIdentifying and moving observations without %s" %uniqueID)
        nullRows = chunk[uniqueID].isnull()
        if nullRows.sum()> 0:
            chunkNA  = chunk[chunk[uniqueID].isnull()]
            print("\t\t\t[%d] rows with NA propid" %(nullRows.sum()))
        #else:
        #    print("\t\tNo observations with a null %s were found" %uniqueID)

        chunk.dropna(subset = [uniqueID], inplace=True) #drop nulls
        chunk.loc[:,uniqueID] = chunk[uniqueID].astype('int')
        chunk.drop_duplicates(subset=[uniqueID],inplace=True)
        
        if histCols is not None:
            chunk = chunk.loc[:,histCols]
        
        # recode columns from bytes to strings
        print("\tRecoding from bytes to strings:")
        chunk.loc[:,chunk.columns[chunk.dtypes=='object']] = chunk.loc[:,chunk.columns[chunk.dtypes=='object']].apply(lambda s: s.str.decode('utf-8'))
        
        print("\tDumping %d observations with a %s to table: %s" %(chunk.shape[0],uniqueID,key))
        
        print("\tQuerying for existing history records...")
        out = queryDB("SELECT transid FROM history WHERE transid IN %s" %str(tuple(chunk.transid)), dbname)
        print("\t\t%d transid records exist in the db already" %out.shape[0])
        print("\t\tdropping non-unique transid observations")
        chunk.drop(chunk.transid.loc[chunk.transid.isin(out.transid)].index, inplace=True)
        
        #open database connection
        print("Connection to %s opened" %dbname)
        conn =  sqlite3.connect("/".join([data_path, dbname]))
        
        try:
            print("\tDumping to db")
            chunk.to_sql(key,conn, if_exists='append', index=False)
            conn.close()
        except sqlite3.Error as er:
            print('\t\t',er)
            print('\t\tskipping chunk')
            conn.close()
        
        chunk_end = time.time()
        print("\t\tdone")
        print("\ttook: %.2f seconds" %((chunk_end-chunk_start)))
        
    end_time = time.time()
    print("this file took: %.2f seconds" %((end_time-start_time)))
        
    #this seems necessary for some reason as data aren't dumped in the end
    del chunk
    del reader
    
    #close database connection
    
    print("done, db connection closed")
```
